```python
Data Structure :
Stack :
Insert/delete always from top.
There is two option push and pop.

class Stack:

    def __init__(self):
        self.values =[] # Used to push or pop.

    def push(self,x):
        return self.values.insert(0,x) # We always insert on 0 position so we use insert.
    
    def pop(self):
        if self.values !=[]:
            return self.values.pop(0)
        else:
            return None # Not want anything in return

    def top(self):
        if self.values !=[]:
            return self.values[0]
        else:
            return None
    
    def isEmpty(self):
        if self.values !=[]:
            return False
        else:
            return True
        
        # if (len(self.values) == 0):
        #      return True
        #  else:
        #      return False
        
    def size(self):
        if (len(self.values)!=0):
            return len(self.values)
        else :
            return "Stack is Empty"
    
    def __str__(self):
        return f"Stack : {self.values}"

l1 = Stack()
print(l1.values)
l1.push(2)
l1.push(3)
l1.push(4)
print(l1)
# l1.pop()
# print(l1.pop())
# print(l1)
print(l1.top())
print(l1.isEmpty())


Queue :

FIFO(First in First Out)
Dequeue : To pop
Queue : To Push

class Queue:

    max_input = 5

    def __init__(self):
        self.values = []

    def enquene(self,x):
        if(len(self.values)<self.max_input):
            return self.values.append(x)
        else:
            return self.isfull()
    
    def dequeue(self):
        if self.values !=[]:
            return self.values.pop(0)
        
        else:
            return None
        
    def peek(self):
        if self.values !=[]:
            return self.values[0]
        else:
            return None
        
    def rear(self):
        if self.values !=[]:
            return self.values[-1]
        else:
            return None
    
    def isEmpty(self):
        if self.values !=[]:
            return False
        else:
            return True
    
    def size(self):
        if (len(self.values)!=0):
            return len(self.values)
        else :
            return "Stack is Empty"
        
    def isfull(self):
        if(len(self.values)>= self.max_input):
            print("Queue is Full!")

    

        
q1 = Queue()
q1.enquene(1)
q1.enquene(2)
q1.enquene(3)
q1.enquene(4)
q1.enquene(5)

print(q1.values)
q1.isfull()
q1.dequeue()
print(q1.values)
print(q1.peek())
print(q1.rear())



```
