```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
x = 2 - 3 * np.random.normal(0, 1, 20)
y = x - 2 * (x ** 2) + 0.5 * (x ** 3) + np.random.normal(-3, 3, 20)
plt.scatter(x,y, s=10)
plt.show()
```

```python
x
```

```python
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
# transforming the data to include another axis
x = x[:, np.newaxis]
y = y[:, np.newaxis]
```

```python
x
```

```python
model = LinearRegression()
model.fit(x, y)
y_pred = model.predict(x)
```

```python
plt.scatter(x, y, s=10)
plt.plot(x, y_pred, color='r')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
rmse = np.sqrt(mean_squared_error(y,y_pred))
r2 = r2_score(y,y_pred)
print(rmse)
print(r2)
```

# The Bias vs Variance trade-off

# Bias refers to the error due to the models simplistic assumptions in fitting the data. A high bias means that the model is unable to capture the patterns in the data and this results in under-fitting.

# Variance refers to the error due to the complex model trying to fit the data. High variance means the model passes through most of the data points and it results in over-fitting the data.

# Regularization

```python
polynomial_features= PolynomialFeatures(degree=30)
```

```python
x_poly = polynomial_features.fit_transform(x)
```

```python
x_poly
```

```python
model = LinearRegression()
model.fit(x_poly, y)
y_poly_pred = model.predict(x_poly)
```

```python
rmse = np.sqrt(mean_squared_error(y,y_poly_pred))
r2 = r2_score(y,y_poly_pred)
print(rmse)
print(r2)
```

```python
import operator
plt.scatter(x, y, s=10)
# sort the values of x before line plot
sort_axis = operator.itemgetter(0)
sorted_zip = sorted(zip(x,y_poly_pred), key=sort_axis)
x, y_poly_pred = zip(*sorted_zip)
plt.plot(x, y_poly_pred, color='m')
plt.show()
```

# Lasso

```python
from sklearn.linear_model import Lasso
```

```python
lasso_1 = Lasso(alpha=10)
```

```python
lasso_1.fit(x_poly, y)
```

```python
lasso_pred = lasso_1.predict(x_poly)
```

```python
lasso_rmse = np.sqrt(mean_squared_error(y,lasso_pred))
print(lasso_rmse)
```

```python
r2_score(y,lasso_pred)
```

```python
sum(lasso_1.coef_ == 0)
```
